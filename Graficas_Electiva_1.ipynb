```python
from matplotlib import legend
import matplotlib.pyplot as plt
import numpy as np

#Grafica de parabola
x = np.linspace( -20, 20, 500)
y = x**2
#el signo de peso al inicio y final deja cursiva la letra
plt.plot(x,y, color = 'red', label = '$f(x) = x^2$', linewidth = 2) # linewidth es para el grosor de la linea azul y el color cambia el color de la linea
plt.title('Función Cuadratica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() #es necesario para el label
plt.grid()
#exportar grafica en cualquier formato
plt.savefig('x2.pdf')
plt.show()

```

```python
from matplotlib import legend
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y, label = '$sen(x)$')
plt.plot(x,y2, label = '$cos(x)$')
plt.title('Funciones Seno y Coseno')

plt.xlabel('x')
plt.ylabel('y')
plt.legend() #es necesario para el label osea la leyenda
plt.grid()
```

```python
 #primero cantidad de filas, segundo cantidad de columnas y tercero en que posicion
plt.subplot(2,1,1)
plt.plot(x,y, "r:", label = "$sen(x)$")
plt.title("Sen(x)")
plt.legend()

plt.subplot(2,1,2)
plt.plot(x,y2,"g--", label = "$cos(x)$")
plt.title("Cos(x)")
plt.legend()

```

Grafica de lineas con estilo cientifico

```python
import matplotlib.pyplot as plt # Librería principal de gráficas
import numpy as np              # Librería para manejo de datos numéricos

# Generamos datos: 10 puntos entre 0 y 10
x = np.linspace(0, 10, 10)
y = x ** 2  # Una parábola simple

# Configuración de la gráfica
plt.figure(figsize=(7, 4)) # Tamaño de la ventana (ancho, alto)

# Graficamos con personalización:
# color: verde, marker: diamante ('d'), linestyle: línea discontinua ('--')
plt.plot(x, y, color='seagreen', marker='d', linestyle='--', linewidth=2, label='Crecimiento')

# Personalización de textos
plt.title('Gráfica de Línea Personalizada', fontsize=14, fontweight='bold')
plt.xlabel('Tiempo (s)')
plt.ylabel('Distancia (m)')
plt.grid(True, alpha=0.3) # Cuadrícula tenue
plt.legend() # Muestra la etiqueta 'Crecimiento'

plt.show() # Muestra el resultado
```

Grafica de barras con colores individuales

```python
import matplotlib.pyplot as plt

# Datos: Categorías y sus valores
frutas = ['Manzanas', 'Bananos', 'Cerezas', 'Uvas']
cantidades = [15, 30, 12, 25]
colores_vivos = ['#FF5733', '#FFC300', '#C70039', '#900C3F'] # Códigos de color Hex

plt.figure(figsize=(7, 4))

# Creamos las barras
# width: ancho de la barra, edgecolor: color del borde
plt.bar(frutas, cantidades, color=colores_vivos, edgecolor='black', width=0.6)

# Añadimos detalles
plt.title('Inventario de Frutas', fontsize=14)
plt.ylabel('Unidades en Stock')
plt.ylim(0, 35) # Forzamos el límite del eje Y para que haya espacio arriba

plt.show()
```

Panel doble, la forma de dividir la pantalla

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. PREPARACIÓN DE DATOS
# Eje X: Representa los meses (del 1 al 5)
meses = np.array([1, 2, 3, 4, 5])

# Eje Y: Aquí definimos qué significan estos números
# Imaginemos que son las ventas de dos productos distintos
ventas_producto_A = [2, 8, 4, 10, 7] # Datos para la gráfica de la izquierda
ventas_producto_B = [10, 5, 12, 3, 8] # Datos para la gráfica de la derecha

# 2. CREACIÓN DEL PANEL (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- GRÁFICA 1 (Izquierda): Ventas de Producto A ---
# Usamos 'fill_between' para crear una gráfica de área
ax1.fill_between(meses, ventas_producto_A, color="skyblue", alpha=0.4)
ax1.plot(meses, ventas_producto_A, color="blue", marker='o') # Línea de contorno
ax1.set_title('Rendimiento: Producto A')
ax1.set_xlabel('Meses del Año')
ax1.set_ylabel('Ventas (en miles $)') # <--- Aquí aclaramos qué es Y
ax1.grid(True, linestyle='--', alpha=0.5)

# --- GRÁFICA 2 (Derecha): Ventas de Producto B ---
# Usamos 'scatter' para ver los puntos aislados (dispersión)
ax2.scatter(meses, ventas_producto_B, color="orange", s=100, edgecolor='red')
ax2.set_title('Rendimiento: Producto B')
ax2.set_xlabel('Meses del Año')
ax2.set_ylabel('Ventas (en miles $)') # <--- Aquí aclaramos qué es Y
ax2.grid(True, linestyle='--', alpha=0.5)

# 3. AJUSTE FINAL
# Evita que las etiquetas de los ejes se amontonen
plt.tight_layout()

# Mostrar el resultado
plt.show()
```
